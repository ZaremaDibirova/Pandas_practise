```python
# Eger musterinin odediyi mebleg 10  dollardan azdirsa $,10 ve 30 arasindadirsa $$,30dan coxdursa $$$ yaz
```

```python
def bill_category(x):
  if x<10:
    return '$'
  elif 10<=x<30:
    return '$$'
  else:
    return '$$$'

df['bill_category']=df['total_bill'].apply(bill_category)
```

##3-cu yol

```python
def bill_category_vector(bill):
  category=np.full_like(bill,fill_value='$',dtype='U3')

  category[(bill>10) & (bill<30)]='$$'

  category[bill>=30]='$$$'

  return category

df['bill_category']=bill_category_vector(df['total_bill'])
df.sample(5)
```
