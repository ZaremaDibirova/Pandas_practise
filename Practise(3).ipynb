```python
# Tapşırıq 18: Əhalisi (Population) 10 milyondan az, lakin ərazisi (Area) 500,000 km²-dən çox olan ölkələri seç və
# onları sahəsinə görə azalan sırada göstər.
```

```python
df.query("`2022 Population`<10000000 and Area>500000").sort_values(by='Area',ascending=False)
```

```python
# Tapsiriq 19: Əhalisinin sıxlığı (Population Density) orta sıxlıqdan çox olan və Area-i median dəyərdən az olan ölkələri tap, 
# onları əhalisinə görə azalan sırada göstər.
```

```python
mean=round(df['Population Density'].mean())
median=round(df['Area'].median())

df.query("`Population Density`>@mean and Area<@median").sort_values(by='2022 Population',ascending=False)
```
