```python
import pandas as pd
```

```python
df=pd.read_csv('info1.csv')
```

```python
# Şəhər  Qəsəbə  Küçə  Xəzər  Bakı  Hövsan  Şüvəlan
```

```python
df.head(3)
```

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.Ad=df.Ad.str.strip()
df['Telefon Nömrəsi']=df['Telefon Nömrəsi'].str.strip()
df['Ailə Vəziyyəti']=df['Ailə Vəziyyəti'].str.strip()
df['İş Statusu']=df['İş Statusu'].str.strip()
```

```python

df[['Şəhər','Rayon','Qəsəbə','Küçə']]=df['Unvan'].str.split(',',expand=True)
df.head(3)
```

```python
df.drop(columns=['Unvan'],inplace=True)
```

```python
df.head(2)
```

```python
df.Yaş=df.Yaş.apply(lambda x: pd.to_datetime('now').year-x if len(str(x))>2 else x)
df.head(2)
```

```python
df['Telefon Nömrəsi']=df['Telefon Nömrəsi'].apply(lambda x: x if x[:4]=='+994' else ('+994 ')+x[:2]+' '+x[2:])
```

```python
df.head(2)
```

```python
# df.loc[df['Telefon Nömrəsi'].str.contains('E'), 'Telefon Nömrəsi'] = 'Məlumat yoxdur'
df['Telefon Nömrəsi'] = df['Telefon Nömrəsi'].apply(lambda x: 'Melumat yoxdur' if 'E' in str(x) else x)
# her iki sekilde yazila biler
```

```python
df.sample(10)
```

```python
%timeit df.loc[df['Telefon Nömrəsi'].str.contains('E'), 'Telefon Nömrəsi'] = 'Məlumat yoxdur'
```

```python
%timeit df['Telefon Nömrəsi'] = df['Telefon Nömrəsi'].apply(lambda x: 'Melumat yoxdur' if 'E' in str(x) else x)
# .apply() metodu daha suretlidir
```

```python
df['Ailə Vəziyyəti']=df['Ailə Vəziyyəti'].str.capitalize()
df['İş Statusu']=df['İş Statusu'].str.capitalize()
```

```python
df.head(2)
```

```python

df.Şəhər=df.Şəhər.replace({'Baku': 'Bakı'})
```

```python
df.head(1)
```

```python
df.Rayon=df.Rayon.str.strip().replace({'Khazar Rayonu':'Xəzər rayonu'})
```

```python
df.head(1)
```

```python
df.Qəsəbə=df.Qəsəbə.str.strip().replace({'Hovsan qəsəbəsi':'Hövsan qəsəbəsi'})
df.head(1)
```

```python
df.Qəsəbə=df.Qəsəbə.str.strip().replace({'Suvelan qəsəbəsi':'Şüvəlan qəsəbəsi'})
df.Qəsəbə=df.Qəsəbə.str.strip().replace({'Mardakan qəsəbəsi':'Mərdəkan qəsəbəsi'})
```
