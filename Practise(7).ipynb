```python
import pandas as pd

# Karmaşık raw veriler
raw_data = [
    'John Doe, 30, "555-1234", "Baku, Azerbaijan", Engineer',
    'Jane Smith, 25, "555-5678", "Ganja, Azerbaijan", Doctor',
    'Hasan Huseynov, 35, "555-9876", "Sumqayit, Azerbaijan", Architect'
]

# Veriyi DataFrame'e yükleme
df = pd.DataFrame(raw_data, columns=["Raw Data"])

# DataFrame'in ilk halinde görüntülenmesi
print(df)
```
